```python
from gensim.models import KeyedVectors
path =  r'/content/gdrive/MyDrive/titans_milestone3/codebase/googlew2v/GoogleNews-vectors-negative300.bin'
word2vec = KeyedVectors.load_word2vec_format(path, binary=True)
```

```python
import numpy as np
from tqdm import tqdm
def getw2v(preprocessed_sent):
    avg_w2v_vectors = []; # the avg-w2v for each sentence/review is stored in this list
    for sentence in tqdm(preprocessed_sent): # for each review/sentence
        vector = np.zeros(300) # as word vectors are of zero length
        cnt_words =0; # num of words with a valid vector in the sentence
        for word in sentence.split(): # for each word in a sentence
            if word in word2vec:
                vector += word2vec[word]
                cnt_words += 1
        if cnt_words != 0:
            vector /= cnt_words
        avg_w2v_vectors.append(vector)
    return avg_w2v_vectors
```

```python
w2v_sent_train=getw2v(preprocessed_sent_train)
w2v_span_train=getw2v(preprocessed_span_train)
```

```python
len(w2v_sent_train[0])
```

```python
w2v_sent_dev=getw2v(preprocessed_sent_dev)
w2v_span_dev=getw2v(preprocessed_span_dev)
```

```python
w2v_sent_test=getw2v(preprocessed_sent_test)
w2v_span_test=getw2v(preprocessed_span_test)
```

```python
from sklearn.preprocessing import StandardScaler
scaler_num = StandardScaler()
X_train_ratio = scaler_num.fit_transform(df_train['ratio'].values.reshape(-1,1))
X_dev_ratio = scaler_num.transform(df_dev['ratio'].values.reshape(-1,1))
X_test_ratio = scaler_num.transform(df_test['ratio'].values.reshape(-1,1))
```

```python
from scipy.sparse import hstack
from scipy.sparse import vstack
X_train = np.hstack((w2v_sent_train,w2v_span_train,X_train_ratio))
X_dev = np.hstack((w2v_sent_dev,w2v_span_dev,X_dev_ratio))
X_test=np.hstack((w2v_sent_test,w2v_span_test,X_test_ratio))
```

```python
y_train=np.asarray(df_train['class'].values)
y_dev=np.asarray(df_dev['class'].values)
```

## logistics regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import SGDClassifier
parameters={'alpha': [0.01,0.1,0.2,0.4,0.6],'penalty':['l1','l2']}
log_reg=SGDClassifier(class_weight = 'balanced', loss='log', random_state=40)
clf = GridSearchCV(log_reg, parameters,scoring='f1_macro',cv=3,return_train_score=True)
clf.fit(X_train, y_train)
```

```python
print("Best Score :",clf.best_score_)
print("Best Estimator :",clf.best_estimator_)
```

```python
clf_best = SGDClassifier(class_weight = 'balanced', alpha=0.01,loss='log', random_state=40)
clf_best.fit(X_train, y_train)
```

```python
y_pred_dev = clf_best.predict(X_dev)
y_pred_test = clf_best.predict(X_test)
```
